```python
import pandas as pd
import geopandas as gpd
```

```python

import folium
from folium import Circle
from folium.plugins import MarkerCluster
```

```python
def embed_map(m, file_name):
    from IPython.display import IFrame
    m.save(file_name)
    return IFrame(file_name, width='100%', height='500px')
```

```python
crimes = pd.read_excel("gasvn.xlsx" )

crimes
```

```python
dt = crimes
dt
```

```python
ten =dt[['ten','dc']]
ten
```

```python
import numpy as np

locations = list(zip(dt.x, dt.y))

ten = list(zip(dt.ten, dt.dc,dt.sdt,dt.img))

popups = ["Tên: {}<br>Địa chỉ: {}<br> Số điện Thoại: {} <br> <img src=' {} '>".format(lat, loc,sdt,img) for (lat,loc,sdt,img) in ten]

from folium.plugins import MarkerCluster

m = folium.Map(
    location=[21,105], tiles="Cartodb Positron", zoom_start=10
)

def color_producer(val):
    if val == 'mở':
        return 'forestgreen'
    else:
        return 'darkred'

for i in range(0,len(dt)):
    Circle(
        location=[dt.iloc[i]['x'], dt.iloc[i]['y']],
        radius=20,
        color=color_producer(dt.iloc[i]['trangthai'])
        ).add_to(m)

marker_cluster = MarkerCluster(
    locations=locations,
    popups=popups,
    name="clustered",

)

marker_cluster.add_to(m)


m
```

```python
embed_map(m, 'index.html')
```
